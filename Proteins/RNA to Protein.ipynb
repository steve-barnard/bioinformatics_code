```python
def rna_to_protein(rna_seq):
    seq_split = [rna_seq[i:i+3] for i in range(0, len(rna_seq), 3)]
    output=[]
    for i in seq_split:
        for k,v in RNA_Codons.items():
            if i == k:
                if v == 'Stop':
                    break
                else:
                    output.append(v)
    return ''.join(output)
```
